# 10 ODI teams

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
data = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

```python
data
```

```python
soup = BeautifulSoup(data.content,"html.parser")
print(soup.prettify())
```

```python
titles = []
for i in soup.find_all("span",class_= "u-show-phablet"):
    titles.append(i.text)
    
titles
```

```python
titles_a = []
for i in soup.find_all("tr", class_="table-body"):
    titles_a.append(i.text.split("\n")[4])
    
titles_a
```

```python
rating = []
for i in soup.find_all("td", class_="table-body__cell u-text-right rating"):
        rating.append(i.text)

    
rating
```

```python
points = []
for i in soup.find_all("tr", class_="table-body"):
    points.append(i.text.split("\n")[8])
    
points
```

```python
matches = []
for i in soup.find_all("tr", class_="table-body"):
    matches.append(i.text.split("\n")[7])
    
matches
```

```python
banner_title = []
for i in soup.find_all("tr", class_="rankings-block__banner"):
    banner_title.append(i.text.split("\n")[4])
    
banner_title
```

```python
banner_rating = []
for i in soup.find_all("tr", class_="rankings-block__banner"):
    banner_rating.append(i.text.split("\n")[10])
    
banner_rating
```

```python
banner_matches = []
for i in soup.find_all("tr", class_="rankings-block__banner"):
    banner_matches.append(i.text.split("\n")[7])
    
banner_matches
```

```python
banner_points = []
for i in soup.find_all("tr", class_="rankings-block__banner"):
    banner_points.append(i.text.split("\n")[8])
    
banner_points
```

```python
df = pd.DataFrame({"titles" :banner_title,"rating":banner_rating,"points":banner_points,"matches":banner_matches})
df
```

```python
df_1 = pd.DataFrame({"titles" :titles_a,"rating":rating,"points":points,"matches":matches})
df_1
```

```python
df_2 = pd.concat([df,df_1],ignore_index = True)
```
