```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
data = requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
data
```

```python
soup = BeautifulSoup(data.content)
soup
```

```python
ft_first = soup.find("h2",class_ = "mp-h2")
ft_first                 
```

```python
ft_first.text
```

```python
titles = []
for i in soup.find_all("h2",class_ = "mp-h2"):
    titles.append(i.text)
    
titles
```

```python
text = []
for i in soup.find_all("div",class_ = "mp-contains-float"):
    text.append(i.text)
    
text
```

```python


image = []
for i in soup.find_all("a",class_ = "image"):
    image.append(i.get("href"))
    
image
```

```python
df = pd.DataFrame({"title" :titles,"text":text,"image":image})
df
```
