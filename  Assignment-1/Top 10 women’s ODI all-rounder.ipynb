```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
data = requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder")
data
```

```python
soup = BeautifulSoup(data.content,"html.parser")
print(soup.prettify())
```

```python
titles_allrounder = []
for i in soup.find_all("td",class_= "table-body__cell rankings-table__name name"):
    titles_allrounder.append(i.text.split("\n")[1])
    
titles_allrounder
```

```python
titles_team = []
for i in soup.find_all("span",class_= "table-body__logo-text"):
    titles_team.append(i.text)
    
titles_team
```

```python
titles_rating = []
for i in soup.find_all("td",class_= "table-body__cell rating"):
    titles_rating.append(i.text)
    
titles_rating
```

```python
name = []
for i in soup.find_all("tr",class_= "rankings-block__banner"):
    name.append(i.text.split("\n")[20])
    
name
```

```python
place = []
for i in soup.find_all("tr",class_= "rankings-block__banner"):
    place.append(i.text.split("\n")[27])
    
place
```

```python
rank = []
for i in soup.find_all("tr",class_= "rankings-block__banner"):
    rank.append(i.text.split("\n")[31])
    
rank
```

```python
df_1 = pd.DataFrame({"NAME" :name,"TEAM":place,"RANK":rank})
df_1
```

```python
df = pd.DataFrame({"NAME" :titles_allrounder,"TEAM":titles_team,"RANK":titles_rating})
df
```

```python
df_2 = pd.concat([df_1,df], ignore_index = True)
df_2
```
