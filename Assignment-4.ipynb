```python

driver = webdriver.Chrome("chromedriver.exe")
time.sleep(3)



url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(3)



most_Viewed_youtube = {}
most_Viewed_youtube['Rank'] = []
most_Viewed_youtube['Name'] = []
most_Viewed_youtube['Artist'] = []
most_Viewed_youtube['Upload_date'] = []
most_Viewed_youtube['Views'] = []
time.sleep(3)


a = 1 



for i in range(0,30):
    time.sleep(3)
    
    try:
        
        rank = driver.find_element(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[1]".format(a))
        most_Viewed_youtube['Rank'].append(rank.text)
        time.sleep(2)
            
        
        name = driver.find_element(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[2]".format(a))
        most_Viewed_youtube['Name'].append(name.text)
        time.sleep(2)
        
        
        artist = driver.find_element(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[3]".format(a))
        most_Viewed_youtube['Artist'].append(artist.text)
        time.sleep(2)
    
        
        uploadDate = driver.find_element(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[5]".format(a))
        most_Viewed_youtube['Upload_date'].append(uploadDate.text)
        time.sleep(2)
    
        
        view = driver.find_element(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[4]".format(a))
        most_Viewed_youtube['Views'].append(view.text)
        time.sleep(2)
    except NoSuchElementException as n:
        print("Exception",n)
   
    # incresing counter variable
    a += 1


Most_Viewed_youtube = pd.DataFrame(most_Viewed_youtube)
Most_Viewed_youtube
```

```python
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(2)


url = "http://statisticstimes.com/"
driver.get(url)
time.sleep(3)

india_gdp = driver.find_element(By.XPATH,"//div[@class='navbar']/div[2]/div[1]/a[3]").get_attribute('href')
driver.get(india_gdp)
time.sleep(3)

state_path = driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[2]/ul/li[1]/a").get_attribute('href')
driver.get(state_path)
time.sleep(3)



gdp_india = {}
gdp_india['Rank'] = []
gdp_india['State'] = []
gdp_india['GSDP at current price (19-20)'] = []
gdp_india['GSDP at current price (18-19)'] = []
gdp_india['Share(18-19)'] =[]
gdp_india['GDP($ billion)'] = []


counter = 1
time.sleep(2)
for i in range(len(driver.find_elements(By.XPATH,"//table[@id='table_id']/tbody/tr"))):
    
    #rank
    rank = driver.find_element(By.XPATH,f"//table[@id='table_id']/tbody/tr[{counter}]/td[1]")
    gdp_india['Rank'].append(rank.text)
    
    state = driver.find_element(By.XPATH,f"//table[@id='table_id']/tbody/tr[{counter}]/td[2]")
    gdp_india['State'].append(state.text)   
               
    #gsdb19_20
    gsdb19_20= driver.find_element(By.XPATH,f"//table[@id='table_id']/tbody/tr[{counter}]/td[3]")
    gdp_india['GSDP at current price (19-20)'].append(gsdb19_20.text)
               
    
    gsdb18_19= driver.find_element(By.XPATH,f"//table[@id='table_id']/tbody/tr[{counter}]/td[4]")
    gdp_india['GSDP at current price (18-19)'].append(gsdb18_19.text)   
    
    share= driver.find_element(By.XPATH,f"//table[@id='table_id']/tbody/tr[{counter}]/td[5]")
    gdp_india['Share(18-19)'].append(share.text)
               
    
    gdb = driver.find_element(By.XPATH,f"//table[@id='table_id']/tbody/tr[{counter}]/td[6]")
    gdp_india['GDP($ billion)'].append(gdb.text)   
    
    counter +=1  # moving to next row
    time.sleep(2)           
               
# Creating Dataframe
Gdp_india = pd.DataFrame(gdp_india)
Gdp_india.to_csv("Gdp_india")
Gdp_india     
```

```python
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(2)


url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
driver.get(url)
time.sleep(3)
```

```python
Novels = {}
Novels["Book_Name"] = []
Novels["Author_Name"] = []
Novels["Volume_Sold"] = []
Novels["Publisher"] = []
Novels["Genre"] = []


count = 1
for i in range(len(driver.find_elements(By.XPATH,"//table[@class='in-article sortable']/tbody/tr"))):
    
    book_name = driver.find_element(By.XPATH,f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[2]")
    Novels["Book_Name"].append(book_name.text)
    
    author_name = driver.find_element(By.XPATH,f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[3]")
    Novels["Author_Name"].append(author_name.text)
    
    volume_name = driver.find_element(By.XPATH,f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[4]")
    Novels["Volume_Sold"].append(volume_name.text)
    
    publisher = driver.find_element(By.XPATH,f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[5]")
    Novels["Publisher"].append(publisher.text)
    
    genre = driver.find_element(By.XPATH,f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[6]")
    Novels["Genre"].append(genre.text)
    
    count += 1
    
Novels_Topselling = pd.DataFrame(Novels)
Novels_Topselling.to_csv("Novels_Topselling.csv")
Novels_Topselling.head(15)
```

```python
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(2)


url = "https://archive.ics.uci.edu/ml/index.php"
driver.get(url)
time.sleep(3)
```

```python
element1 = driver.find_element(By.XPATH,"/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b")
element1.click() 
```

```python
uci_repositories = {}
uci_repositories['Dataset name']= []
uci_repositories['Data type'] = []
uci_repositories['Task'] = []
uci_repositories['Attribute type'] = []
uci_repositories['No of instances'] = []
uci_repositories['No of attribute'] = []
uci_repositories['Year'] = []

name = driver.find_elements(By.XPATH,"//p[@class='normal']/b/a")
for i in name:
    if len(uci_repositories['Dataset name']) < 586:
        uci_repositories['Dataset name'].append(i.text)
    else:
        pass
     
time.sleep(1)
for i in range(2,588):
    dtype = driver.find_element(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[2]/p")
    if dtype.text == " ":
        uci_repositories['Data type'].append("-")
    else:
        uci_repositories['Data type'].append(dtype.text)
    
    task = driver.find_element(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[3]/p")
    if task.text == " ":
        uci_repositories['Task'].append("-")
    else:
        uci_repositories['Task'].append(task.text) 
        
    type = driver.find_element(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[4]/p")
    if type.text == " ":
        uci_repositories['Attribute type'].append("-")
    else:
        uci_repositories['Attribute type'].append(type.text)
        
    instance = driver.find_element(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[5]/p")
    if instance.text == " ":
        uci_repositories['No of instances'].append("-")
    else:
        uci_repositories['No of instances'].append(instance.text) 
        
    num_attribute = driver.find_element(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[6]/p")
    if num_attribute.text == " ":
        uci_repositories['No of attribute'].append("-")
    else:
        uci_repositories['No of attribute'].append(num_attribute.text)        
        
    year = driver.find_element(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[7]/p")
    if year.text == " ":
        uci_repositories['Year'].append("-")
    else:
        uci_repositories['Year'].append(year.text)
time.sleep(1)


#Creating Dataframe
UCI_Repositories = pd.DataFrame(uci_repositories)
UCI_Repositories.to_csv("UCI_Repositories.csv")
UCI_Repositories.head(20)
```
