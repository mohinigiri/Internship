```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
data = requests.get("https://presidentofindia.nic.in/former-presidents.htm")
```

```python
data
```

```python
soup = BeautifulSoup(data.content,"html.parser")
print(soup.prettify())
```

```python
titles = []
for i in soup.find_all("div",class_ = "presidentListing"):
    titles.append(i.text.split('\n')[1].split('(')[0])
    
titles
```

```python
date = []
for i in soup.find_all("div",class_ = "presidentListing"):
    date.append(i.text.split('\n')[2].split(":")[1])
    
date
```

```python
df = pd.DataFrame({"titles":titles,"date":date})
df
```
