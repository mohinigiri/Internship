```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
data = requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling")
data
```

```python
soup = BeautifulSoup(data.content,"html.parser")
print(soup.prettify())
```

```python
titles_balling = []
for i in soup.find_all("td",class_= "table-body__cell rankings-table__name name"):
    titles_balling.append(i.text.split("\n")[1])
    
titles_balling
```

```python
titles_ballingteam = []
for i in soup.find_all("span",class_= "table-body__logo-text"):
    titles_ballingteam.append(i.text)
    
titles_ballingteam
```

```python
titles_ballingrating = []
for i in soup.find_all("td",class_= "table-body__cell rating"):
    titles_ballingrating.append(i.text)
    
titles_ballingrating
```

```python
name = []
for i in soup.find_all("tr",class_= "rankings-block__banner"):
    name.append(i.text.split("\n")[20])
    
name
```

```python
place = []
for i in soup.find_all("tr",class_= "rankings-block__banner"):
    place.append(i.text.split("\n")[27])
    
place
```

```python
rank = []
for i in soup.find_all("tr",class_= "rankings-block__banner"):
    rank.append(i.text.split("\n")[31])
    
rank
```

```python
df_1 = pd.DataFrame({"NAME" :name,"TEAM":place,"RANK":rank})
df_1
```

```python
df = pd.DataFrame({"NAME" :titles_balling,"TEAM":titles_ballingteam,"RANK":titles_ballingrating})
df
```

```python
df_2 = pd.concat([df_1,df], ignore_index = True)
```
